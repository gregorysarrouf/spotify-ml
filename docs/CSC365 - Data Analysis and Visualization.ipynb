## Plotting on axes objects

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

A common issue in Matplotlib is overlapping subplots or figure elements like titles, labels, or tick marks. To resolve this, the **fig.tight_layout() or plt.tight_layout()** method can be used. It automatically adjusts subplot spacing to ensure all elements are displayed clearly without overlap. This method recalculates axis positions, making it especially helpful for complex visualizations with multiple subplots, where manual adjustments can be tedious.

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)  

plt.tight_layout()
```

## Parameters on subplots()

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[0][0].set_title('Title[0][0]')


axes[1][1].plot(x,y)
axes[1][1].set_title('Title[1][1]')
axes[1][1].set_xlabel('X Label[1][1]')

axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.suptitle("Figure Title",fontsize=16)


plt.show()
```

## Exporting plt.subplots()

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('subplots.png',bbox_inches='tight')

plt.show()
```

## Legends

Toadd legends to a figure by using the **label="label text"** keyword argument while creating plots or other objects. Once the labels are defined, simply call the legend() method without any arguments to display the legend on the figure.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

You can determine the location of the legend using the argument: **loc**

Check http://matplotlib.org/users/legend_guide.html#legend-location) for more details. 

```python
# ax.legend(loc=1) # upper right corner
# ax.legend(loc=2) # upper left corner
# ax.legend(loc=3) # lower left corner
# ax.legend(loc=4) # lower right corner

# # # loc=0: matplotlib will give the optimal location
ax.legend(loc=0) 
fig
```

```python
ax.legend(loc=(1.1,0.5)) # manually set location
fig
```

## Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, color = 'blue') 
ax.plot(x, x**3, color = 'green')
```

```python
# Or
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

## Line and marker styles

## Linewidth

We can set the line width using `linewidth` or `lw` keyword argument. 

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use linewidth or lw
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10)
```

## Linestyles

We can set the line style using **linestyle** argument

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
```

## Markers

We can define markers shape and size using **marker** and **markersize** arguments.

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use marker for string code
# Use markersize or ms for size

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20)
```

## displot() and histplot()  
The displot is a versatile plot type used to visualize the distribution of a single feature. It primarily displays a histogram, with an optional overlay of a Kernel Density Estimation (KDE) curve to provide a smooth representation of the data distribution. 

```python
sns.displot(data=df,x='salary',kde=True)
```
